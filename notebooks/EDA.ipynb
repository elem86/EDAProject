```python
import pandas as pd
import plotly.express as px
import numpy as np

df = pd.read_csv("/Programming/EDAProject/EDAProject/vehicles_us.csv")


df.describe()
```

```python
df.info()
#We havesome datatypes that need to be converted.
```

```python
df.head()
```

```python
df.isna().sum()
```

```python
df.groupby('is_4wd')['is_4wd'].count()

# Wanted to see the values in the column
```

```python
df['is_4wd'] = df['is_4wd'].fillna(0).astype(bool)
df.groupby('is_4wd')['is_4wd'].count()

# First I filled the NaN values with 0 so I can calculate with the column later on. Changed the type of the column to int from float.
```

```python
df["model_year"] = df["model_year"].fillna(df.groupby('model')["model_year"].transform('median'))
df["cylinders"] = df["cylinders"].fillna(df.groupby('model')["cylinders"].transform('median'))
df["odometer"] = df["odometer"].fillna(df.groupby('model_year')["odometer"].transform('median'))

# Replaced the missing values in the numerical colomuns with the median values and changed the datatypes to int.
df.isna().sum()
```

```python
df['odometer'] = df['odometer'].fillna(0)
# found a row where the odometer was 0, probably a new car, this is the fix I applied. Since it is one row of data, there is no significance to it. 

types_to_replace = ['model_year', 'cylinders', 'odometer']
for column in types_to_replace:
    df[column] = df[column].apply(np.int64)

df.isna().sum()    
```

```python
df['paint_color'] = df['paint_color'].fillna("unknown")
# paint_color was the only non numberic column with mising values, replaced those values with "unknown".
```

```python
df['date_posted'] = df['date_posted'].astype('datetime64[ns]')
# Converted the date_posted column to datetime type.
```

```python
df.info()
```

```python
df.isna().sum()
```

```python
df.duplicated().sum()
# No duplicates in the dataset.
```

```python
df['manufacturer'] = df['model'].apply(lambda x: x.split()[0])
df.head()

# I think it is a good idea to create a new column with the manufacturer only.
```

```python
fig = px.histogram(df, x="model_year", nbins=25, histnorm='density',                     
                    color_discrete_sequence=["blue"],
                    barmode='overlay',
                    title='Distribution of Cars by Model Year')
fig.update_layout(xaxis={'categoryorder': 'total descending'})
fig.update_traces(opacity=0.7)
fig.show()

# Distribution of Cars by Condition can provide good information about the market.
```

```python
fig = px.histogram(df, x="manufacturer",
                   histnorm='density', 
                   color_discrete_sequence=["blue"],
                   barmode='overlay',
                   title='Distribution of Cars by Manufacturer')
fig.update_layout(xaxis={'categoryorder': 'total descending'})
fig.update_traces(opacity=0.7)
fig.show()

# Distribution of Cars by Manufacturer shows the amount of ads per manufacturer.
```

```python


df_avg_price = df.groupby(['manufacturer', 'model_year'])['price'].mean().reset_index()

colors = ['#1f77b4', '#ff7f0e', '#2ca02c', '#d62728', '#9467bd', '#8c564b', '#e377c2', '#7f7f7f', '#bcbd22', '#17becf', '#20B2AA', '#7B68EE', '#FFA07A']

fig = px.scatter(df_avg_price, x='model_year', y='price', color='manufacturer',
             title='Average Price by Manufacturer and Year',
             color_discrete_sequence=colors,
             labels={'year': 'Year', 'price': 'Average Price', 'manufacturer': 'Manufacturer'})
fig.show()

# Average Price by Manufacturer and Year scatter plot gives a solid idea about it there is a relation between car manufacturer, the age of the  car and the average price.
```

```python
fig = px.scatter(df, x='model_year', y='days_listed',
             title='Year vs. days listed',
             labels={'model_year': 'Year', 'days_listed': 'Days Listed'})
fig.show()


# Year vs. days listed chart showing us that the oldsmobile and new cars are on the market less.
```

# FINAL CONCLUSION

The goal of this data research is to find out which factors influence the price of a vehicle. 

The dataframe contains data of 51525 advertisements.

The dataset contained missing values in several columns, 5 to be exact.

The missing values in the model_year column were replaced based on the median values of the specific model of the car from the model column.

The missing values in the cylinders column were replaced based on the median values of the same cylinders column grouped by the model.

Paint_color column's missing values were replaced with 'unknown' since these values are in fact unknown.

Is_4wd column is an interesting one. The datatype is set to numerical values so Python thought the 0s are missing values, these has been replaced with actual 0s and the datatype corrected to boolean.

In the odometer column I have replaced the missing values with the median values based on the model_year column since there is a direct correlation between the car's age and the mileage.

The dataset has no duplicate values.

Created 2 histograms that are organized by descending order used by the plotly.express library. Also created two scatter plots using the same library.

The overall conclusion is this short analysis is that the most ads are from Ford, oldsmobiles are selling fast and their price is getting higher. Newer cars are most expensive than and older car but not oldsmobile car yet.
